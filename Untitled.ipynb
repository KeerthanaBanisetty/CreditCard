```python
avg_inc=data["Annual_income"].mean()
avg_inc
```

```python
gender_income=data.groupby("GENDER")["Annual_income"].mean()
gender_income
```

```python
gender_income.sort_values(ascending=True)
```

```python
sns.barplot(x=gender_income.index,y=gender_income.values)
plt.title("Average by Gender_income")
plt.show()
```

```python
label_count=data["label"].value_counts()
label_count
```

```python
plt.pie(label_count,labels=["Credit Approved","Credit Rejected"],autopct="%1.1f%%")
plt.title("Credit Approval Status")
plt.show()
```
